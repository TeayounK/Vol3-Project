```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load in data
data = pd.read_csv("Life Expectancy Data Cleaned With Insurance.csv")
```

```python
# Split data between where it does and doesn't have universal healthcare
mask1 = data["Universal_insurance"] == True
mask2 = data["Universal_insurance"] == False
uni_info = data[mask1]
other_info = data[mask2]
# Get the mean life expectancy of each portion
uni_mean = uni_info["Life_expectancy"].mean()
other_mean = other_info["Life_expectancy"].mean()
print("Mean of universal healthcare life expectancy\n",uni_mean)
print("Mean of non universal healthcare life expectancy\n",other_mean)
```

```python
# Make box and whisker plots of each section
fig, axes = plt.subplots(1, 2, sharey = True)
boxplot1 = uni_info.boxplot(column=["Life_expectancy"], ax = axes[0])
axes[0].set_title("Universal Healthcare")
axes[0].set_xticklabels([])
boxplot2 = other_info.boxplot(column=["Life_expectancy"], ax = axes[1])
axes[1].set_title("Non Universal Healthcare")
axes[1].set_xticklabels([])
axes[0].set_ylabel("Life Expectency")
plt.tight_layout()
plt.savefig("Box and Whisker I Barely Even Know Her.png", dpi = 300)
```
